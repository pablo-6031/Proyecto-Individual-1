# Carga de datos

```python
import pandas as pd

# Cargar el dataset
credit = pd.read_csv("credits.csv")
movies = pd.read_csv("movies_dataset.csv")

```

# TRANSFORMACIONES

# Desanidados de datos

```python
import ast
# Convertir la columna 'belongs_to_collection' en diccionarios si están en formato de string
movies['belongs_to_collection'] = movies['belongs_to_collection'].apply(
    lambda x: ast.literal_eval(x) if isinstance(x, str) else x
)

# Extraer el nombre de la colección si existe y guardarlo en una nueva columna
collection_name= movies['belongs_to_collection'].apply(
    lambda x: x['name'] if isinstance(x, dict) and 'name' in x else None
)

collection_name
```

```python
# Convertir la columna 'genres' en listas de diccionarios si están en formato de string
movies['genres'] = movies['genres'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else [])

# Extraer solo los nombres de los géneros y unirlos en una sola cadena separada por comas
genres_names = movies['genres'].apply(lambda x: ', '.join([genre['name'] for genre in x if isinstance(genre, dict) and 'name' in genre]))

genres_names
```

```python

# Convertir la columna 'production_companies' en listas de diccionarios si están en formato de string
movies['production_companies'] = movies['production_companies'].apply(
    lambda x: ast.literal_eval(x) if isinstance(x, str) else []
)

# Extraer solo los nombres de las compañías y unirlos en una sola cadena separada por comas
production_companies_names = movies['production_companies'].apply(
    lambda x: ', '.join([company['name'] for company in x if isinstance(company, dict) and 'name' in company]) if isinstance(x, list) else ""
)

production_companies_names
```

```python

# Convertir la columna 'production_companies' en listas de diccionarios si están en formato de string
movies['production_countries'] = movies['production_countries'].apply(
    lambda x: ast.literal_eval(x) if isinstance(x, str) else []
)

# Extraer solo los nombres de las compañías y unirlos en una sola cadena separada por comas
production_countries_names = movies['production_countries'].apply(
    lambda x: ', '.join([company['name'] for company in x if isinstance(company, dict) and 'name' in company]) if isinstance(x, list) else ""
)

production_countries_names
```

```python
# Convertir la columna 'spoken_languages' en listas de diccionarios si están en formato de string
movies['spoken_languages'] = movies['spoken_languages'].apply(
    lambda x: ast.literal_eval(x) if isinstance(x, str) else []
)

# Extraer solo los nombres de los lenguajes y unirlos en una sola cadena separada por comas
spoken_languages_names = movies['spoken_languages'].apply(
    lambda x: ', '.join([company['name'] for company in x if isinstance(company, dict) and 'name' in company]) if isinstance(x, list) else ""
)

spoken_languages_names
```

Archivo Credits

```python
# Convertir la columna 'cast' en listas de diccionarios si están en formato de string
credit['cast'] = credit['cast'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else [])

# Extraer solo los nombres de los actores y unirlos en una sola cadena separada por comas
credit['cast'] = credit['cast'].apply(lambda x: ', '.join([genre['name'] for genre in x[:3] if isinstance(genre, dict) and 'name' in genre]))


credit['cast']
```

```python
# Convertir la columna 'cast' en listas de diccionarios si están en formato de string
credit['crew'] = credit['crew'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else [])


# Función para extraer el valor de "name" de los registros con "job" igual a "Director"
def extraer_director(var):
    for v in var:
        if v['job'] == 'Director':
            return v['name']
    return None


# Extraer solo los nombres del director 
credit['name_director'] = credit['crew'].apply(lambda x: extraer_director(x))


```

```python
credit.head()
```

```python
credit.drop('crew', axis=1, inplace=True)
```

```python
movies['belongs_to_collection'] = collection_name
movies['genres'] = genres_names
movies['production_companies'] = production_companies_names
movies['production_countries'] = production_countries_names
movies['spoken_languages'] = spoken_languages_names

movies
```

```python
# Convertir la columna a int64
movies = movies[movies['id'].str.isdigit()]

credit['id'] = credit['id'].astype('int64')
movies['id'] = movies['id'].astype('int64')
```

```python
movies.info()
```

```python


# Combinar los DataFrames en función de la columna 'id'
movies_full = pd.merge(movies, credit, on='id', how='inner')
```

```python
movies_full
```

# Valores nulos

```python
nulos_columna = movies_full['release_date'].isna().sum()

print("Valores nulos ", nulos_columna)
```

```python
# Asegurarse de que 'revenue' y 'budget' no tengan valores nulos y sean numéricos
movies_full['revenue'] = pd.to_numeric(movies_full['revenue'], errors='coerce').fillna(0)
movies_full['budget'] = pd.to_numeric(movies_full['budget'], errors='coerce').fillna(0)
```

```python
movies_full = movies_full.dropna(subset=['release_date'])

# Convertir 'release_date' al formato de fecha AAAA-mm-dd
movies_full['release_date'] = pd.to_datetime(movies_full['release_date'], errors='coerce').dt.strftime('%Y-%m-%d')
```

# Fecha

```python
print(movies_full['release_date'].head(10))
```

```python
# Crear la columna 'release_year' extrayendo solo el año
movies_full['release_year'] = pd.to_datetime(movies_full['release_date'], errors='coerce').dt.year
```

# Retorno de inversion

```python

# Calcular el retorno de inversión (revenue / budget) y asignar 0 donde no es posible calcular
movies_full['return'] = movies_full.apply(lambda x: x['revenue'] / x['budget'] if x['budget'] > 0 else 0, axis=1)

# Mostrar las primeras filas para verificar el cálculo
print(movies_full[['title', 'revenue', 'budget', 'return']].head(10))
```

# Eliminacion de columnas

```python
# Eliminar las columnas que no serán utilizadas
movies_full = movies_full.drop(columns=['video', 'imdb_id', 'adult', 'original_title', 'poster_path', 'homepage'], errors='ignore')

# Mostrar las primeras filas para verificar que las columnas se han eliminado
print(movies_full.head())
```

```python

# Exportar el DataFrame a un archivo CSV
# Filtrar las películas lanzadas después de 1990 al 2000
movies_full = movies_full[(movies_full['release_year'] > 1990) & (movies_full['release_year'] < 2000)]
movies_full.to_csv("movies_transform.csv", index=False)
```
